```python
import numpy as np
import random
from munkres import Munkres, print_matrix

numberObjectif=4#nombre d'objectifs
#lecture
with open('LAP-30.dat') as f:
    size = int(f.readline()) #taille de la matrice
    alldata = [] #toute les données des objectifs
    for k in range(numberObjectif):
        data = []
        for i in range(size):
            data.append(f.readline())
        for i in range(len(data)):
            data[i]=data[i].strip('\n').split()
            for j in range(len(data[i])):
                data[i][j]=int(data[i][j])
        alldata.append(data)

```

```python
def update(newResult,newSolution ,allResult,allSolution):  #ici newResult représente un resultat et non tout le vecteur de nouveau resultat
    globalCount=0
    vecpop= []

    for i in range(len(allResult)):
        countPetit=0 # conteur de plus petit pour chaque objectif (old < new)
        countGrand=0 #conteur de plus grand plur chaque onjectif  (old > new)
        for j in range(len(allResult[i])):
            if(allResult[i][j]<= newResult[j]):
                countPetit+=1
            if(allResult[i][j]>=newResult[j]):
                countGrand+=1

        if(countPetit==len(newResult)):
            break # stop car la nouvelle solution se fait dominer
        if(countGrand==len(newResult)):
            #allResult.pop(i)
            vecpop.append(i) # enregistre l'élément qui se fait dominer

            #allResult.append(newResult)

        globalCount +=1

    decalage = 0

    for i in range(len(vecpop)):
        allResult.pop(vecpop[i]-decalage)  #supression de l'élément
        allSolution.pop(vecpop[i]-decalage) #supression de l'élément
        decalage = decalage +1
        
    if(globalCount -len(vecpop) == len(allResult)):
        allResult.append(newResult)
        #maodifier
        allSolution.append(newSolution)
    
    return allResult, allSolution


```

```python
allSolution = [] #vecteur avec tout les solutions (vecteur de 0 et de 1)
allResult = [] #Vecteur avec tout les résultats de chaque solution 

#solution exacte
for i in range(len(alldata)):
    m = Munkres()
    indexes = m.compute(alldata[i])
    test = np.zeros((size,size))
    for i in indexes:
        test[i[0],i[1]]=1
   
    result = calculResult(alldata,test)
    
    allResult,allSolution = update(result,test,allResult,allSolution)
    

#création des solutions initiale (diagonale)
solutionDiagonal = np.zeros((size,size))
for i in range(len(solutionDiagonal)):
    solutionDiagonal[i][i]=1

result = calculResult(alldata, solutionDiagonal)
allResult,allSolution = update(result,solutionDiagonal,allResult,allSolution)

#création de solution diagonal inverse
solutionDiagonalInverse = np.flip(solutionDiagonal,0)

result = calculResult(alldata,solutionDiagonalInverse)
allResult,allSolution = update(result,solutionDiagonalInverse,allResult,allSolution)

#création d'une matrice alétoire
nombreSolutionAleatoire=100
while(len(allSolution)<7):#for j in range(nombreSolutionAleatoire):
    li=[]
    for i in range(size):
        li.append(i)
    se=set(li)
    li=list(se)
    ordre = random.sample(li,len(li))
    solutionrandom = np.zeros((size,size))
    for i in range(len(solutionrandom)):
        solutionrandom[i][ordre[i]]=1
    
    result = calculResult(alldata, solutionrandom)
    allResult,allSolution = update(result,solutionrandom,allResult,allSolution)



print(allResult)



####  regarder domination
#### créer plus de alétoire
#### ajouter toutes les solutions dans un vecteur de solution
```
